# Lab Courses Over Time

The output is a graph where the Y-axis is the number of courses with labs and the X-axis is the year. \
The graph is a line graph showing the # of lab courses per year. \

```python
import matplotlib.pyplot as plt
from utils.get_all_dfs import * 

semesters = get_all_dfs("courses")
semester_labels = list(semesters.keys())

sem_lab_section_counts = []
for semester in semesters:
    # Count number of rows (courses) with lab_instructor != NaN
    sem_lab_section_counts.append(semesters[semester].lab_instructor.count())

lab_section_counts_yearly = [sem_lab_section_counts[i] + sem_lab_section_counts[i+1] for i in range(0, len(sem_lab_section_counts), 2) if i+1 < len(sem_lab_section_counts)]
year_labels = list(map(lambda label: label[0:4], semester_labels[::2])) #Every other semester, first 4 chars
if len(year_labels) % 2 != 0:
    year_labels.pop()
    
# total_courses = [len(semesters[semester]) for semester in semesters]
# num_courses_yearly = [total_courses[i] + total_courses[i+1] for i in range(0, len(total_courses), 2) if i+1 < len(total_courses)]
# if len(num_courses_yearly) % 2 != 0:
#     num_courses_yearly.pop()
    
# print(num_courses_yearly)
    
# labs_per_course = [lab_section_counts_yearly[i] / num_courses_yearly[i] for i in range(len(lab_section_counts_yearly))]


plt.style.use('seaborn-v0_8-darkgrid')
plt.title("Number of Courses w/ Labs Over Time")
plt.xlabel("Year")
plt.ylabel("Total Number of Courses (in Spring +  Fall) w/ Labs")
plt.xticks(rotation=90)
plt.plot(year_labels, lab_section_counts_yearly)
```
