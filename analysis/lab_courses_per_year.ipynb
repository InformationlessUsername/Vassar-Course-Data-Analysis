# Lab Courses per Year

The output is a graph where the Y-axis is the number of courses with labs and the X-axis is the number of years

```python
import matplotlib.pyplot as plt
from utils.get_all_dfs import *

years = get_all_dfs("courses", by_year=True)
```

```python
# Format of {year: count, ...} 
years_labs = {}

for year, courses in years.items():
    # Skip 2024 because it's incomplete
    if year == "2024": continue 
        
    # Add year to possible years
    possible_years.append(year)
        
    for index, course in courses.iterrows():
        lab_course = type(course["lab_instructor"]) == str
        # Skip non-lab courses
        if not lab_course: continue

        # Add year to years_labs if it doesn't exist
        if year not in years_labs:
            years_labs[year] = 0
        
        # Increment count for year
        years_labs[year] += 1

plt.title("Number of Lab Courses per Year")
plt.xlabel("Year")
plt.ylabel("Number of Lab Courses")
plt.xticks(rotation=90)
plt.plot(years_labs.keys(), years_labs.values(), label=dept)
```
